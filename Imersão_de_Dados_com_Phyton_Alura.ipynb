```python
# Mapeamento das colunas em inglês para português
traducao_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Traduzir as colunas do DataFrame
df = df.rename(columns=traducao_colunas)

# Exibir as novas colunas para confirmar a tradução
print(df.columns)
```

```python
df['senioridade'].value_counts()
#com base na biblioteca, o .value_counts()  irá puxar a freq de cada categoria/valor que está dentro da coluna

#SE = senior
#MI = mid (intermediário, pleno)
#EN = entry (iniciante, junior)
#EX = executivo
```

```python
df['remoto'].value_counts()
#0 trabalho presencial
#100 trabalho remoto
#50 trabalho híbrido
```

```python
df['tamanho_empresa'].value_counts()
#M medium
#L large
#S small
```

```python
# Mapeamento das categorias de senioridade em inglês para português
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

# Traduzir as categorias na coluna 'senioridade'
df['senioridade'] = df['senioridade'].replace(senioridade)

# Exibir a contagem das novas categorias para confirmar a tradução
df['senioridade'].value_counts()
```

```python
# Mapeamento das categorias de contrato em inglês para português
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freela'
}

# Traduzir as categorias na coluna 'contrato'
df['contrato'] = df['contrato'].replace(contrato)

# Exibir a contagem das novas categorias para confirmar a tradução
df['contrato'].value_counts()
```

```python
# Mapeamento das categorias de tamanho de empresa em inglês para português
traducao_tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

# Traduzir as categorias na coluna 'tamanho_empresa'
df['tamanho_empresa'] = df['tamanho_empresa'].replace(traducao_tamanho_empresa)

# Exibir a contagem das novas categorias para confirmar a tradução
display(df['tamanho_empresa'].value_counts())
```

```python
# Mapeamento das categorias de remoto para português
traducao_remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

# Traduzir as categorias na coluna 'remoto'
df['remoto'] = df['remoto'].replace(traducao_remoto)

# Exibir a contagem das novas categorias para confirmar a tradução
display(df['remoto'].value_counts())
```

##Exemplos fictícios para testar preenchimento de valores nulos


> Add blockquote



### Muitas vezes quando os valores são nulos e queremos preenchê-las podemos usar uma estratégia de usa da ***média*** ou da ***mediana*** dos valores para preencher esse vazio

```python
import numpy as np
#numpy é uma biblioteca

df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#preencher com a média salarial
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#preencher com a mediana salarial
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
#fillna significa fill numero nulo

df_salarios
#usa esse df_salario no final para chamar na hora de exibir
```

###Exemplo de preenchimento com o ***valor anterior ou posterior***, em alguns casos, como temperatura, não convém usar a media ou a mediana

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
#ffill = forward fill
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
#bfill = backward fill
df_temperaturas
```

###Exemplo de preenchimento com ***valor fixo***[link text](https://)

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

Analisando a distribuição dos salários


*   Histograma

```python
#HISTOGRAMA

plt.figure(figsize=(8,4))
#figsize (largura, altura)

sns.histplot(df_limpo['usd'], bins=50, kde=True)
#bins define o intervalo entre as barras geradas pelo histograma
#kde é a linha usada para entender como está a distribuição, se colocar false ela não aparece

plt.title('Distribuição dos salários anuais')
plt.xlabel('Salario em USD')
plt.ylabel('Frequência')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='senioridade')
#configurando as cores do boxplot, palette - define as cores | hue - parâmetro
#que designa uma cor para cada categoria da variável
plt.title("Boxplot da distribuição por senioridade")
plt.xlabel('Salário em USD')
plt.ylabel('Nível de Senioridade')
plt.show()
```

##Criação de gráficos interativos com a biblioteca Plotly

```python
import plotly.express as px
```

```python
#propmt: crie um gráfico de média salarial por senioridade em barras usando o plotly

# Calcular a média salarial por senioridade
media_salario_por_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()

# Ordenar os dados pela média salarial
media_salario_por_senioridade = media_salario_por_senioridade.sort_values('usd', ascending=False)

# Criar o gráfico de barras interativo com Plotly Express
fig = px.bar(media_salario_por_senioridade,
             x='senioridade',
             y='usd',
             title='Média Salarial por Nível de Senioridade',
             labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial Anual (USD)'})

# Exibir o gráfico
fig.show()
```

```python
# Criar o gráfico de pizza interativo com Plotly Express
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie (remoto_contagem,
              names = 'tipo_trabalho',
              values = 'quantidade',
              title = 'Proporção dos tipos de trabalho')

fig.show()
```

```python
# Criar o gráfico de pizza/rosca interativo com Plotly Express
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie (remoto_contagem,
              names = 'tipo_trabalho',
              values = 'quantidade',
              title = 'Proporção dos tipos de trabalho',
              hole=0.5)

fig.update_traces(textinfo='percent+label')
fig.show()
```
